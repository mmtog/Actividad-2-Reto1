```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv sin indice
df = pd.read_csv('inmuebles2.csv')
df
```

```python
#Eliminar primera fila o filas innecesarias
df1= pd.read_csv('inmuebles2.csv', skiprows=1)
df1
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
df2= df1.drop(['0','Referencia'], axis=1)
df2
```

```python
#Registros fecha alta
filtro1=df2[(df2["Fecha Alta"]=="2007-04-29 00:00:00") & (df2["Fecha Alta"]=="2004-01-02 00:00:00")]
filtro1
```

No existen registros con esa fecha

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("filtro1.csv")
```

```python
filtro2 = df2[(df2["Tipo"]=="Parking") | (df2["Tipo"]=="Local") | (df2["Tipo"]=="Industrial") | (df2["Tipo"]=="Piso")]
filtro2
```

```python
filtro2.to_csv("filtro2.csv")
```

```python
#Filtro3. registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
filtro3= df2[(df2["Provincia"]=="Giona") & (df2["Superficie"]> 100)]
filtro3
```

```python
filtro3.to_csv("filtro3.csv")
```

Análisis filtro3: No se puede cumplir con la solicitud porque se emplea la función & , en cambio, si utilizaramos | , sí arrojaría los datos de la tabla.

```python
# Filtro4. registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”
filtro4= df2[(df2["Precio Venta"]>1000000) | (df2["Vendedor"]=="Carmen") | (df2["Vendedor"]=="Joaquin") ]
filtro4
```

```python
filtro4.to_csv("filtro4.csv")
```

```python
# Filtro5. registros que no sean de “Provincia : Barcelona“
filtro5= df2[~(df2["Provincia"]=="Barcelona")]
filtro5
```

```python
filtro5.to_csv("filtro5.csv")
```

```python
#Filtro6. registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
filtro6=df2[(df2["Operación"]=="Alquiler") &(df2["Tipo"]=="Oficina")]
filtro6
```

```python
filtro6.to_csv("filtro6.csv")
```
